```python
import pandas as pd
```

```python
dataDF = pd.read_excel(r'HDB1026/AMIRS.xlsx', 'AMIRS')
```

```python
# Hard coded IRS-Shibor code
codes = ['AMIRSRMB03MA5Q3SH',
         'AMIRSRMB06MA5Q3SH',
         'AMIRSRMB09MA5Q3SH',
         'AMIRSRMB01YA5Q3SH',
         'AMIRSRMB02YA5Q3SH',
         'AMIRSRMB03YA5Q3SH',
         'AMIRSRMB04YA5Q3SH',
         'AMIRSRMB05YA5Q3SH',
         'AMIRSRMB07YA5Q3SH',
         'AMIRSRMB10YA5Q3SH']
```

```python
shiborIRSQuote =  dataDF[(dataDF['RIC NAME'] == codes[0]) | (dataDF['RIC NAME'] == codes[1])|
       (dataDF['RIC NAME'] == codes[2]) | (dataDF['RIC NAME'] == codes[3])|
       (dataDF['RIC NAME'] == codes[4]) | (dataDF['RIC NAME'] == codes[5])|
       (dataDF['RIC NAME'] == codes[6]) | (dataDF['RIC NAME'] == codes[7])|
       (dataDF['RIC NAME'] == codes[8]) | (dataDF['RIC NAME'] == codes[9])]
shiborIRSQuote.sort(['Asia/Shanghai DATE ', 'Asia/Shanghai TIME '], inplace = True)
import datetime as dt
shiborIRSQuote = shiborIRSQuote[(shiborIRSQuote['Asia/Shanghai TIME '] <= dt.time(17,30,0)) & 
               (shiborIRSQuote['Asia/Shanghai TIME '] >= dt.time(8,0,0))]
```

```python
dailyAverage = shiborIRSQuote.groupby(['Asia/Shanghai DATE ', 'RIC NAME']).mean()
```

```python
dailyAverage['MID'] = (dailyAverage['ASK'] + dailyAverage['BID']) / 2.0
```

```python
dailyAverage = dailyAverage.unstack()
```

```python
midPrice = dailyAverage['MID']
```

```python
columns = midPrice.columns
```

```python
newColumnNames = []
for name in columns:
    newColumnNames.append(name[8:11].strip('0'))
```

```python
midPrice.columns = newColumnNames
```

```python
midPrice.dropna(inplace = True)
midPrice = midPrice.reindex(columns = ['6M', '9M', '1Y', '2Y', '3Y', '4Y', '5Y', '7Y', '10Y'], inplace= True)
```

```python
dateIndex = midPrice.index
columns = midPrice.columns
```

```python
from CAL import *
totalRates = []
tenors = ['6M', '9M', '1Y', '2Y', '3Y', '4Y', '5Y', '6Y', '7Y', '8Y', '9Y', '10Y']
shiborIndex = Shibor(Period(3, Months))
dc = ActualActual(ActualActual.ISMA)
calendar = China(China.IB)
```

```python
for index in dateIndex:
    timeStamp = pd.Timestamp(index)
    calDate = Date(timeStamp.day, timeStamp.month, timeStamp.year)
    record = midPrice.ix[index]
    Settings.instance().setEvaluationDate(calDate)
    rateHelpers = []
    for col in columns:
        rate = record[col] / 100.0
        if not pd.isnull(rate):
            tenor = PeriodParser.parse(str(col))
            quote = SimpleQuote(rate)
            quoteHandler = QuoteHandle(quote)
            fixFreq = Semiannual
            
            inst = ShiborSwapRateHelper(quoteHandler, tenor, fixFreq, shiborIndex)
            rateHelpers.append(inst)
    
    
    
    if calendar.isBusinessDay(calDate):
        curve = PiecewiseFlatForward(calDate, rateHelpers, dc)

        dates = [calDate + Period(x) for x in tenors]
        rates = {}
        rates['date'] = index
        for tenor, date in zip(tenors,dates):
            rate = curve.zeroRate(date, dc, Compounded, Annual)
            rates[tenor] = rate.rate()
        totalRates.append(rates)
    else:
        pass
        
```

```python
df = pd.DataFrame(totalRates)
df.reindex(columns = ['date', '6M', '9M', '1Y', '2Y', '3Y', '4Y', '5Y', '6Y', '7Y', '8Y', '9Y', '10Y']).sort('date')
```

```python
df.to_excel(u'Shibor互换.xlsx', cols= ['date', '6M', '9M', '1Y', '2Y', '3Y', '4Y', '5Y', '6Y', '7Y', '8Y', '9Y', '10Y'])
```
