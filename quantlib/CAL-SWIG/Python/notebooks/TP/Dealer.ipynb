```python
import pandas as pd
from CAL import *
```

```python
df = pd.read_excel(u'同花顺报价数据.xlsx','Sheet1')
treasuryInfo = pd.read_excel(u'国债数据.xls', 'Sheet1')
```

```python
ootrBonds = df[(df['alias'] == 'OO1Y') |
   (df['alias'] == 'OO3Y') |
   (df['alias'] == 'OO5Y') |
   (df['alias'] == 'OO7Y') |
   (df['alias'] == 'OO10Y')|
   (df['alias'] == 'OO30Y')]
```

```python
treasuryInfo[:5]
treasuryInfo = treasuryInfo[['BOND_NAME', 'TICKER_SYMBOL', 'FIRST_ACCRUAL_DT', 'ISSUE_PRICE', 'ISSUE_DT', 'EXCHANGE_LISTED_DT','COUPON_TYPE_CD','CPN_FREQ_CD', 'COUPON', 'MATURITY_DT']]
treasuryInfo = treasuryInfo[treasuryInfo["COUPON_TYPE_CD"] != 'FLOAT']
treasuryInfo.dropna(subset = ['EXCHANGE_LISTED_DT'], inplace = True)
def changex(x):
    try:
        return int(x)
    except:
        return x
    
def stringToDate(x):
    try:
        return pd.Timestamp(str(x))
    except:
        return x
    
treasuryInfo['TICKER_SYMBOL'] = treasuryInfo['TICKER_SYMBOL'].apply(changex)
```

```python
def fillEmptyFreq(x):
    if x == '1PY':
        return 1
    elif x == '2PY':
        return 2
    elif x == 'MAT':
        return 1
    else:
        return 0
```

```python
res = pd.merge(ootrBonds, treasuryInfo, how = 'left', left_on='ticker', right_on='TICKER_SYMBOL')
res['PAY_FREQ'] = res['CPN_FREQ_CD'].apply(fillEmptyFreq)
res['COUPN_RATE'] = res['COUPON']
```

```python
res['quoteDate'] = res.date
```

```python
res = res.sort('quoteDate')
```

```python
res[:6]
```

```python
def getTenor(x):
    if len(x)==3 and x[0] == 'O' and x[:2] != 'OO':
        return x[-2:]
    elif len(x) == 4 and x[0] == 'O'and x[:2] != 'OO':
        return x[-3:]
    elif len(x) == 4 and x[:2] == 'OO':
        return x[-2:]
    elif len(x) == 5 and x[:2] == 'OO':
        return x[-3:]
    else:
        return x
```

```python
res['tenor'] = res.alias.apply(getTenor)
```

```python
res = res[['ticker', 'alias', 'quoteDate', 'tenor', u'均价', 'FIRST_ACCRUAL_DT', 'MATURITY_DT', 'COUPN_RATE', 'PAY_FREQ']].sort(['quoteDate', 'MATURITY_DT'])
```

```python
res.to_excel(u'做市商曲线基准券.xlsx')
```

```python
groupRes = res.groupby('quoteDate')
settlementDays = 1
dc = ActualActual(ActualActual.ISMA)
yieldCurves = {}
totalRates = []
```

```python
 for key in groupRes.indices:
    date = key
    tmp = key.date()
    evalDate = Date(tmp.day, tmp.month, tmp.year)
    Settings.instance().setEvaluationDate(evalDate)
    ratesHelper = []
    instruments = groupRes.get_group(key)
    index = instruments.index
    for inst in index:
        tmp = instruments.ix[inst]['tenor']
        tenor = PeriodParser.parse(str(tmp))
        maturityDate = instruments.ix[inst]['MATURITY_DT']
        maturityDate = Date(maturityDate.day, maturityDate.month, maturityDate.year)
        startDate = maturityDate - tenor
        coupon = instruments.ix[inst]['COUPN_RATE'] / 100.0
        cleanPrice = instruments.ix[inst][u'均价']
        # Build bond rate helper based on clean price
        quote = SimpleQuote(cleanPrice)
        quoteHandler = QuoteHandle(quote)
        
        if instruments.ix[inst]['PAY_FREQ'] == 1:
            frequency = Annual
        elif instruments.ix[inst]['PAY_FREQ'] == 2:
            frequency = Semiannual
        else:
            raise Exception("")
            
        tenor = Period(frequency)
        bondHelper = CTBFixedBondHelper(quoteHandler,
                                            settlementDays,
                                            100.0,
                                            startDate,
                                            maturityDate,
                                            tenor,
                                            [coupon],
                                            dc)
            
        ratesHelper.append(bondHelper)
    yieldCurve = PiecewiseFlatForward(evalDate,
                                        ratesHelper,
                                        dc)
    
    yieldCurves[evalDate] = yieldCurve
        
    # Calculate benchmark curve points
    tenors = [str(x) + 'Y' for x in xrange(1,10)]
    dates = [evalDate + Period(x) for x in tenors]
    rates = {}
    rates['date'] = date
    for tenor, date in zip(tenors,dates):
        rate = yieldCurve.zeroRate(date, dc, Compounded, Annual, True)
        rates[tenor] = rate.rate()
        
    totalRates.append(rates)
```

```python
resDF = pd.DataFrame(totalRates)
```

```python
resDF = resDF.reindex(columns = ['date'] + tenors).sort('date')
```
