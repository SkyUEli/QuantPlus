```python
import pandas as pd
from CAL import *
import datetime as dt
```

```python
treasuryDF = pd.read_excel(u'国债报价.xls', 'Sheet1')
```

```python
otrBonds = pd.read_excel(u'OTR序列.xls', 'Sheet1')
```

```python
treasuryInfo = pd.read_excel(u'国债数据.xls', 'Sheet1')
```

```python
def changex(x):
    try:
        return int(x)
    except:
        return x
    
def stringToDate(x):
    try:
        return pd.Timestamp(str(x))
    except:
        return x

def fillEmptyFreq(x):
    if x == 1 or x == 2:
        return x
    else:
        return 0
    
def totalMonthToPeriod(x):
    tmp = x / 12
    if tmp > 0:
        return str(tmp) + 'Y'
    else:
        return str(x) + 'M'
```

```python
treasuryDF = treasuryDF[['OFFC_CODE2', 'Time', 'COUPN_RATE', 'PAY_FREQ', 'YIELD', 'MATUR_DATE']]
treasuryInfo = treasuryInfo[['BOND_NAME', 'TICKER_SYMBOL', 'FIRST_ACCRUAL_DT', 'ISSUE_PRICE', 'ISSUE_DT', 'EXCHANGE_LISTED_DT','COUPON_TYPE_CD', 'MATURITY_DT']]
treasuryInfo = treasuryInfo[treasuryInfo["COUPON_TYPE_CD"] != 'FLOAT']
treasuryInfo.dropna(subset = ['EXCHANGE_LISTED_DT'], inplace = True)
    
treasuryInfo['TICKER_SYMBOL'] = treasuryInfo['TICKER_SYMBOL'].apply(changex)
treasuryDF['MATUR_DATE'] = treasuryDF['MATUR_DATE'].apply(stringToDate)
    
# Attache tenor info to treasury info data frame
treasuryInfo['YEAR'] = treasuryInfo['MATURITY_DT'].apply(lambda x: x.year) - treasuryInfo['FIRST_ACCRUAL_DT'].apply(lambda x: x.year) 
treasuryInfo['MONTH'] = treasuryInfo['MATURITY_DT'].apply(lambda x: x.month) - treasuryInfo['FIRST_ACCRUAL_DT'].apply(lambda x: x.month) 
treasuryInfo['TOTAL_MONTH'] = treasuryInfo['YEAR'] * 12 + treasuryInfo['MONTH']
    
treasuryInfo['TENOR'] = treasuryInfo['TOTAL_MONTH'].apply(totalMonthToPeriod)
treasuryInfo = treasuryInfo[['BOND_NAME', 'TICKER_SYMBOL', 'FIRST_ACCRUAL_DT', 'ISSUE_PRICE', 'ISSUE_DT', 'EXCHANGE_LISTED_DT','COUPON_TYPE_CD', 'MATURITY_DT',
                             'TENOR']]
```

```python
calendar = China(China.IB)
startDate = Date(16,12, 2013)
endDate = Date(31,3,2014)

workingDates = []
while startDate<=endDate:
    if calendar.isBusinessDay(startDate):
        workingDates.append(startDate)
    startDate = startDate + Period(1, Days)
```

```python
treasuryDF.sort(['Time', 'MATUR_DATE'], inplace = True)
```

```python
treasuryDF = pd.merge(treasuryDF, treasuryInfo[['TICKER_SYMBOL', 'TENOR']], left_on = ['OFFC_CODE2'], right_on=['TICKER_SYMBOL'])
```

```python
treasuryDF['QUOTE_DATE'] = treasuryDF['Time'].apply(lambda x: x.date())
```

```python
treasuryDF['alias'] = ''
```

```python
indexList = treasuryDF.index
for index in indexList:
    record = treasuryDF.ix[index]
    quoteDate = record['QUOTE_DATE']
    otrRecord = otrBonds.ix[quoteDate]
    loop = otrRecord.index
    for lIndex in loop:
        if otrRecord[lIndex]!='None':
            if record['OFFC_CODE2'] == otrRecord[lIndex]:
                record['alias'] = lIndex
    
    treasuryDF.ix[index] = record
    
```

```python
treasuryDF.to_excel(u'国债报价_扩展.xlsx')
```

```python
longestDatespan = 5
timeInterval = dt.timedelta(hours = 12)
tenorPoints = ['1Y', '3Y', '5Y', '7Y', '10Y', '30Y']
patternString = '|'.join(tenorPoints + [''])
wokringDF = treasuryDF[treasuryDF['alias'].str.contains(patternString)]

columns = treasuryDF.columns
insruments = []

for date in workingDates:
    startTime = dt.datetime(date.year(),date.month(),date.dayOfMonth(),9,0)
    endTime = dt.datetime(date.year(),date.month(),date.dayOfMonth(),21,0,0)
    timePoints = []
    while startTime <= endTime:
        timePoints.append(startTime)
        startTime = startTime + timeInterval
        
    longestWorkingDates = calendar.advance(date, -longestDatespan, Days)
    pyDate = dt.datetime(date.year(),date.month(),date.dayOfMonth())
    pyPreviousDate = dt.datetime(longestWorkingDates.year(),longestWorkingDates.month(),longestWorkingDates.dayOfMonth(),0,0,0)
    otrList = otrBonds.ix[pyDate]
    for timePoint in timePoints:
        usedCodes = set()
        for tenor in tenorPoints:
            codes = [otrList['OO' + tenor]]#, otrList['OO' + tenor]
            codes = set(codes)
            for code in codes:
                choosed = wokringDF[(wokringDF['OFFC_CODE2'] == code) &
                                    (wokringDF['Time']<= timePoint) &
                                    (wokringDF['Time']>= pyPreviousDate)]
                if len(choosed)!=0:
                    res = choosed[-1:]
                    record = {}
                    record['CurveTime'] = timePoint
                    for col in columns:
                        record[col] = res[col].values[0]
                    insruments.append(record)
                    usedCodes.add(code)
                    
        moreBondsToConsider = set()
        moreBondsToConsider = moreBondsToConsider - usedCodes
        for code in moreBondsToConsider:
            choosed = wokringDF[(wokringDF['OFFC_CODE2'] == code) &
                                    (wokringDF['Time']<= timePoint) &
                                    (wokringDF['Time']>= pyPreviousDate)]
            if len(choosed)!=0:
                res = choosed[-1:]
                record = {}
                record['CurveTime'] = timePoint
                for col in columns:
                    record[col] = res[col].values[0]
                insruments.append(record)
```

```python
bigRes = pd.DataFrame(insruments)
```

```python
bigRes = bigRes[['CurveTime', 'alias', 'OFFC_CODE2', 'Time', 'MATUR_DATE', 'PAY_FREQ', 'COUPN_RATE', 'TENOR', 'YIELD']]
```

```python
bigRes.sort(['CurveTime', 'MATUR_DATE'], inplace = True)
bigRes.to_excel(u'实时收益率曲线基准券.xlsx')
```

```python
index = bigRes.groupby('CurveTime').count().index
```

```python
settlementDays = 1

tenors = []
for i in xrange(1,21):
    tenors.append(str(i) + 'Y')
dc = ActualActual(ActualActual.ISMA)
totalRates = []
for i in index:
    dayDF = bigRes[bigRes['CurveTime'] == i]
    innerIndex = dayDF.index
    date = i.date()
    evalDate = Date(date.day, date.month, date.year)
    dates = [evalDate + Period(x) for x in tenors]
    Settings.instance().setEvaluationDate(evalDate)
    
    ratesHelper = []
    
    for inst in innerIndex:
            tmp = dayDF.ix[inst]['TENOR']
            tenor = PeriodParser.parse(str(tmp))
            maturityDate = dayDF.ix[inst]['MATUR_DATE']
            maturityDate = Date(maturityDate.day, maturityDate.month, maturityDate.year)
            startDate = maturityDate - tenor
            coupon = dayDF.ix[inst]['COUPN_RATE'] / 100.0
            yieldOfBond = dayDF.ix[inst]['YIELD'] / 100.0
            
            timeToMaturity = (maturityDate - evalDate) / 365.0
            if timeToMaturity < 1.0:
                compounding = Simple
            else:
                compounding = Compounded
            frequency = dayDF.ix[inst]['PAY_FREQ']
            
            if frequency == 1:
                frequency = Annual
            elif frequency == 2:
                frequency = Semiannual
            else:
                raise Exception("")
            
            tenor = Period(frequency)
            
            # Start build the bond and convert yield to clean price
            bondObj = CTBFixedBond(issueDate = startDate,
                                   settlementDays = settlementDays,
                                   faceAmount = 100.0,
                                   startDate = startDate,
                                   maturity = maturityDate,
                                   tenor = tenor,
                                   coupons = [coupon],
                                   accrualDayCounter = dc)
            
            cleanPrice = bondObj.cleanPrice(yieldOfBond,
                                            dc, 
                                            compounding,
                                            frequency)
            
            # Build bond rate helper based on clean price
            quote = SimpleQuote(cleanPrice)
            quoteHandler = QuoteHandle(quote)
            
            bondHelper = CTBFixedBondHelper(quoteHandler,
                                            settlementDays,
                                            100.0,
                                            startDate,
                                            maturityDate,
                                            tenor,
                                            [coupon],
                                            dc)
            
            ratesHelper.append(bondHelper)
            
    yieldCurve = PiecewiseFlatForward(evalDate, ratesHelper, dc)
        
    # Calculate benchmark curve points
    dates = [evalDate + Period(x) for x in tenors]
    rates = {}
    rates['quote'] = i
    for tenor, date in zip(tenors,dates):
        rate = yieldCurve.zeroRate(date, dc, Compounded, Annual, True)
        rates[tenor] = rate.rate()
        
    totalRates.append(rates)
```

```python
df = pd.DataFrame(totalRates).reindex(columns = ['quote'] + tenors)
df.to_excel(u'实时收益率曲线.xlsx')
```
