## 1. 如何获取市场日线数据？使用MarketDataGet！

```python
from CAL.PyCAL import *
import pandas as pd
defaultFigSize = (15,15)
defaultFontSize = 20
```

```python
res =  MarketDataGetHistDailyQuotes(securityID = "000002.XSHE", 
                                    fields = ["shortNM", "closePrice", "turnoverValue", "capitalInflow"], 
                                    startDate = "20000106", 
                                    endDate = "20140110")
```

```python
res[:5]
```

```python
Plot(res, settings = {'x':'tradeDate','y':'closePrice', 'title':u'万科历史收盘价格', 'figsize':defaultFigSize,
                      'fontsize':defaultFontSize})
```

```python
closePrice = res['closePrice']
dailyReturn = closePrice[:-1].values / closePrice[1:] - 1
res['return'] = dailyReturn
```

```python
res[:10]
```

```python
Plot(res, settings = {'x':'closePrice', 'y':'return', 'kind':'Scatter', 'title':u'价格 vs. 收益率（散点图', 
                      'figsize':defaultFigSize, 'fontsize':defaultFontSize})
```

## 2. 使用向量函数计算万科股票虚拟在值看涨期权

```python
priceList = BSMPrice(optionType = 1, 
                     strike = res['closePrice'] + 1.0, 
                     spot = res['closePrice'], 
                     riskFree = 0.00, 
                     dividend = 0.00, 
                     volatility = 0.25, 
                     maturity = 2.0)
priceList['strike'] = res['closePrice'] + 1.0
priceList['stockPrice'] = res['closePrice']
```

```python
priceList[:10]
```

```python
Plot(priceList, settings = {'y':['price'], "title":u"万科期权价格 2010~2014", 'figsize':defaultFigSize, 
                            'fontsize':defaultFontSize})
```

## 3. 获取财报数据？使用FundamentalBalanceSheetRange

```python
res = FundamentalBalanceSheetRange(securityID = '600110',
                                   startDate = '20050101',
                                   endDate = '20140101',
                                   reportType = 'Q')
```

```python
res[:10]
```

```python
Plot(res, settings= {'x':u'END_DATE', 'y':[u'TOTAL_ASSETS',u'TOTAL_LIAB', u'TOTAL_EQUITY'], 
                     'kind':'Bar', 'title':u'浦发银行2007~2013季度报表', 'figsize':defaultFigSize,
                     'fontsize':defaultFontSize})
```

## 4. 获取股票分类静态数据

```python
res = SecurityCategory(standard = 'SW', level = 1)
```

```python
countDF = res.groupby('industryName')
```

```python
res[:10]
```

```python
ticker = pd.DataFrame(countDF['ticker'].count())
```

```python
ticker.columns = ['No. of Corp.']
ticker[:10]
```

```python
Plot(ticker, settings = {'x':'No. of Corp.', 'kind':'Pie','title':u'申万一级行业分类统计', 
                         'figsize':defaultFigSize,
                         'fontsize':defaultFontSize})
```
