```python
import pandas as pd
from matplotlib import pylab
import datetime as dt
from math import *
from scipy.stats import skew
import numpy as np
import pprint
defaultFigSize = (12,10)
```

```python
class PerformanceEvaluator:
    def __init__(self, orderBook, startBalance = 500.0, commissionRate = 0.0, slippage = 2, slipSize = 0.01 ):
        self.order_book = orderBook.copy()
        self.order_book['openDate'] = orderBook.index.get_level_values(0)
        self.order_book['openTime'] = orderBook.index.get_level_values(1)
        self.start_balance = startBalance
        self.commission_rate = commissionRate
        self.num_trades = len(self.order_book)
        self.slippage = slippage
        self.slip_size = slipSize
        
        # Calculate year fraction
        time_index =  self.order_book['openDate'].values
        time_delta = np.double((time_index[-1] - time_index[0]) / 24.0/ 3600 / 1000000000) / 365.0
        self.year_fraction = time_delta
        self.__calculate_order_detail__()
    
    def __calculate_order_detail__(self):
        self.order_book['commission'] = np.abs(self.order_book['share']) * (self.order_book['closePrice'] + self.order_book['openPrice']) * self.commission_rate
        self.order_book['cum_commission'] = self.order_book['commission'].cumsum()
        self.total_commission = self.order_book['cum_commission'][-1:].values[0]
        self.order_book['slippage_cost'] = np.abs(self.order_book['share']) * 2.0 * self.slippage * self.slip_size
        self.order_book['cum_slippage_cost'] = self.order_book['slippage_cost'].cumsum()
        self.total_slippage_cost = self.order_book['cum_slippage_cost'][-1:].values[0]
        self.order_book['raw_profit'] = self.order_book['share'] * (self.order_book['closePrice'] - self.order_book['openPrice'])
        self.order_book['cum_raw_profit'] = self.order_book['raw_profit'].cumsum()
        self.total_raw_profit = self.order_book['cum_raw_profit'][-1:].values[0]
        self.order_book['profit'] = self.order_book['raw_profit'] - self.order_book['commission'] - self.order_book['slippage_cost']  
        self.order_book['cum_profit'] = self.order_book['profit'].cumsum()
        self.total_profit = self.order_book['cum_profit'][-1:].values[0]
        self.order_book['win?'] = self.order_book['profit'] > 0
        tmp = self.order_book['profit']
        absProfit = np.maximum (tmp,0)
        absLoss = np.minimum(tmp,0)
        self.order_book['gross_win'] = absProfit.cumsum()
        self.gross_win = self.order_book['gross_win'][-1:].values[0]
        self.order_book['gross_loss'] = absLoss.cumsum()
        self.gross_loss = self.order_book['gross_loss'][-1:].values[0]
        self.order_book['win_trades'] = np.cumsum(self.order_book['win?'] == True)
        self.win_trades = self.order_book['win_trades'][-1:].values[0]
        self.order_book['loss_trades'] = np.cumsum(self.order_book['win?'] == False)
        self.loss_trades = self.order_book['loss_trades'][-1:].values[0]
        self.order_book['average_profit_per'] = self.order_book['cum_profit'] / (self.order_book['win_trades'] + self.order_book['loss_trades']).astype('double')
        self.average_profit_per = self.order_book['average_profit_per'][-1:].values[0]
        self.order_book['average_win_per'] = self.order_book['gross_win'] / self.order_book['win_trades'].astype('double')
        self.average_win_per = self.order_book['average_win_per'][-1:].values[0]
        self.order_book['average_loss_per'] = self.order_book['gross_loss'] / self.order_book['loss_trades'].astype('double')
        self.average_loss_per = self.order_book['average_loss_per'][-1:].values[0]
        self.order_book['win_loss_ratio'] = self.order_book['average_win_per'] / (-self.order_book['average_loss_per'])
        self.win_loss_ratio = self.order_book['win_loss_ratio'][-1:].values[0]
        self.order_book['winning_percentage'] = self.order_book['win_trades'] / (self.order_book['win_trades'] + self.order_book['loss_trades'])
        self.winning_percentage = self.order_book['winning_percentage'][-1:].values[0]
        self.order_book['profit_factor'] = self.order_book['gross_win'] / (-self.order_book['gross_loss'])
        self.profit_factor = self.order_book['profit_factor'][-1:].values[0]
        self.order_book['return'] = self.order_book['profit'] / self.start_balance
        self.order_book['cum_return'] = self.order_book['cum_profit'] / self.start_balance
        self.total_return = self.order_book['cum_return'][-1:].values[0]
        self.order_book['cum_balance'] = self.order_book['cum_profit'] + self.start_balance
        self.end_balance = self.order_book['cum_balance'][-1:].values[0]
        
        # Calculate some fancy indicator       
        draw_down_res = self.__calculate_max_drawdown__(self.order_book['cum_return'].values)
        index = self.order_book.index.get_level_values(0)
        self.max_draw_down = draw_down_res['max_draw_down']
        draw_down_vec = draw_down_res['draw_down_vec']
        self.average_draw_down = np.mean(draw_down_vec)
        self.draw_down_duration = np.double(index[draw_down_res['draw_down_low_index']].values - index[draw_down_res['draw_down_high_index']].values)/ 24.0/ 3600 / 1000000000
        self.max_draw_down_duration = self.draw_down_duration.max()
        self.average_draw_down_duration = self.draw_down_duration.mean()
        
        best_trades_index = self.order_book['profit'].argmax()
        worst_trades_index = self.order_book['profit'].argmin()
        returnCol = ['share', 'openDate', 'openTime', 'openPrice', 'closeDate', 'closeTime', 'closePrice', 'win?', 'raw_profit', 
                     'commission', 'slippage_cost', 'profit']
        self.best_trade = self.order_book.ix[best_trades_index][returnCol]
        self.worst_trade = self.order_book.ix[worst_trades_index][returnCol]
        
        # return distribution
        self.__calculate_return_stats__(self.order_book[['openDate', 'cum_balance']])
        
        # calculate daily stats
        self.daily_trades_count = self.__calculate_day_stats__(self.order_book)
        self.average_trades_daily = self.daily_trades_count.mean()
    
    def __calculate_day_stats__(self, order_book):
        res = order_book.groupby('openDate').count()
        res = res['share']
        res.name = 'trades_count'
        return res
    
    def __calculate_return_stats__(self, cum_balance_vec):
        res = cum_balance_vec.groupby('openDate').last()
        self.daily_balance = res
        tmp = np.append([self.start_balance], res['cum_balance'][:-1].values)
        res['daily_return'] = (res['cum_balance'] / tmp - 1.0) * 252.0
        self.daily_return = res['daily_return']
        self.return_mean = res['daily_return'].mean()
        self.return_skweness = skew(res['daily_return'].values)
        self.return_volatility = std(res['daily_return'].values)
        self.sharp_ratio = self.return_mean / self.return_volatility
        self.downside_return_volatility = std(res[res['daily_return'] <= 0]['daily_return'].values)
        self.sortino_ratio = self.return_mean / self.downside_return_volatility
    
    def __calculate_max_drawdown__(self, cum_profit_vec):
        running_low = cum_profit_vec[0]
        running_high = cum_profit_vec[0]
        
        running_max_drawdonw = 0.0
        low_index = 0
        high_index = 0
        lookback_max = cum_profit_vec[0]
        lookback_max_index = 0
        
        draw_down_vec = []
        draw_down_low = []
        draw_down_high = []
        
        for j, cum_profit in enumerate(cum_profit_vec):
            running_low = cum_profit_vec[j]
            running_high = cum_profit_vec[j]
            drawdown = min(running_low - lookback_max, 0.0)
            draw_down_vec.append(drawdown)
            draw_down_low.append(j)
            draw_down_high.append(lookback_max_index)
            if drawdown < running_max_drawdonw:
                low_index = j
                high_index = lookback_max_index
                running_max_drawdonw = drawdown
            if running_high > lookback_max:
                lookback_max_index = j
                lookback_max = running_high
                    
        return {'max_draw_down': running_max_drawdonw,
                'max_low_index':low_index,
                'max_high_index':high_index,
                'draw_down_vec':draw_down_vec,
                'draw_down_low_index':draw_down_low,
                'draw_down_high_index':draw_down_high}
```

```python
parse = lambda x: dt.datetime.strptime(str(int(x)), '%Y%m%d')
data = pd.read_csv('SPY_tradeList_1min.txt',  parse_dates = [0,4], index_col = [0,1], date_parser = parse)
per_eval = PerformanceEvaluator(data, startBalance = 200.0, commissionRate = 0.0, slippage = 0, slipSize = 0.01)
```

```python
def formatString(fieldName, fieldValue):
    return fieldName + "," + str(fieldValue) + "\n"
try:
    file = open('Performance.dat', 'w')
        
    # Main stats on portfolio
    file.writelines(formatString("Start Balance", per_eval.start_balance))
    file.writelines(formatString("Year Fraction", per_eval.year_fraction))
    file.writelines(formatString("Commission Rate", per_eval.commission_rate))
    file.writelines(formatString("Slippage", per_eval.slippage))
    file.writelines(formatString("Min Tick Size ", per_eval.slip_size))
    file.writelines(formatString("End Balance", per_eval.end_balance))
    file.writelines(formatString("Total Raw Profit", per_eval.total_raw_profit))
    file.writelines(formatString("Total Commission Cost", per_eval.total_commission))
    file.writelines(formatString("Total Slippage Cost", per_eval.total_slippage_cost))
    file.writelines(formatString("Total Profit", per_eval.total_profit))
    file.writelines(formatString("Total Return", per_eval.total_return))
    file.writelines(formatString("Average Return(annual)", per_eval.return_mean))
    file.writelines(formatString("Return Vol(annual)", per_eval.return_volatility))
    file.writelines(formatString("Sharp ratio", per_eval.sharp_ratio))
    file.writelines(formatString("Skewness(annual)", per_eval.return_skweness))
        
    # Some performance result
    file.writelines(formatString("Max Draw Down", per_eval.max_draw_down))
    file.writelines(formatString("Max Draw Down Duration(days)", per_eval.max_draw_down_duration))
    file.writelines(formatString("Average Draw Down", per_eval.average_draw_down))
    file.writelines(formatString("Average Draw Down Duration(days)", per_eval.average_draw_down_duration))
        
    # detail on trades level
    file.writelines(formatString("Gross Win", per_eval.gross_win))
    file.writelines(formatString("Gross Loss", per_eval.gross_loss))
    file.writelines(formatString("Profit Factor", per_eval.profit_factor))
    file.writelines(formatString("Total Trades", per_eval.num_trades))
    file.writelines(formatString("Win Trades", per_eval.win_trades))
    file.writelines(formatString("Loss Trades", per_eval.loss_trades))
    file.writelines(formatString("Winning Percentage", per_eval.winning_percentage))
    file.writelines(formatString("Average Number Trades(per day)", per_eval.average_trades_daily))
    file.writelines(formatString("Average Profit(per trade)", per_eval.average_profit_per))
    file.writelines(formatString("Average Win(per trade)", per_eval.average_win_per))
    file.writelines(formatString("Average Loss(per trade)", per_eval.average_loss_per))
    file.writelines(formatString("Win-Loss ratio(per trade)", per_eval.win_loss_ratio))
    file.writelines(formatString("Best Winning Trade", per_eval.best_trade))
    file.writelines(formatString("Worst Lossing Trade", per_eval.worst_trade))
        
finally:
    file.close()
```
