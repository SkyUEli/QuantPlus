```python
sql = 'SELECT * FROM chinabond.cb_bond_tnc'
bondInfo = psql.frame_query(sql, con=conn) 
```

```python
res = df[['xibe_ticker', 'notional_value']].groupby('xibe_ticker').sum()
bondInfo = bondInfo.set_index('xibe_ticker')
```

```python
liqudityReport = pd.merge(res, bondInfo[[ 'actual_issue_amt']], left_index='xibe_ticker', right_index='xibe_ticker', how='left')
```

```python
liqudityReport['turn_over'] = liqudityReport['notional_value'] /liqudityReport['actual_issue_amt']
```
